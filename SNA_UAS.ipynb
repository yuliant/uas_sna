```python
import pandas as pd

# Read the Excel file into a dataframe
dataframe = pd.read_excel('/content/uas_sna/data_olahraga.xlsx')
```

```python
import networkx as nx

graph2 = nx.Graph()

for index, row in dataframe.iterrows():
    siswax2 = row['NRP']
    sisway2 = row[1:]

    graph2.add_node(siswax2)

    for i, siswa2 in enumerate(sisway2):
        if siswa2>= 1:
            siswa_name = dataframe.columns[i + 1]
            graph2.add_node(siswa_name)
            graph2.add_edge(siswax2, siswa_name,weight=siswa2)

print(f"Number of nodes: {graph2.number_of_nodes()}")
print(f"Number of edges: {graph2.number_of_edges()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a dictionary of edge labels with weights
edge_labels = {(u, v): weight for u, v, weight in graph2.edges.data('weight')}

plt.figure(figsize=(12, 8))  # Adjust the dimensions as desired
pos = nx.kamada_kawai_layout(graph2)

nx.draw_networkx_nodes(graph2, pos, node_color='lightblue', node_size=300)
nx.draw_networkx_edges(graph2, pos, edge_color='gray', width=1, alpha=0.5)
nx.draw_networkx_labels(graph2, pos, font_size=10)
nx.draw_networkx_edge_labels(graph2, pos, edge_labels=edge_labels, font_size=8)

plt.axis('off')
plt.tight_layout()

plt.show()
```

# Nomer 8
Temukan apakah Graph ini memenuhi paradigma Friendship Paradox

```python
import networkx as nx

# Assuming you already have the graph 'graph'

# Calculate the average degree of the nodes
degree_nodes = dict(graph.degree())
avg_degree_nodes = sum(degree_nodes.values()) / len(degree_nodes)

# Calculate the average degree of the neighbors
avg_degree_neighbors = 0
for node in graph.nodes():
    neighbors = graph.neighbors(node)
    avg_degree_neighbors += sum(degree_nodes[n] for n in neighbors)
avg_degree_neighbors /= len(graph)

# Check if the friendship paradox holds
if avg_degree_nodes < avg_degree_neighbors:
    print("Friendship Paradox holds")
else:
    print("Friendship Paradox does not hold")
```
